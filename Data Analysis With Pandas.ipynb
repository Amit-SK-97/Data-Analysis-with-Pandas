# Working with dates

```python
df.date
```

The date series is a object datatype which should be converted into a date format for further analysis

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df.date
```

The datatype has been changed into datetime64

It has been done so as to extract data based on year, month and so on. Had it been an object dataype, these functionalities would require 

For loop

String Manipulation

Appending it back to the data frame

But pandas does all that by just converting the datecolumn using .to_datetime()

Creating new columns of Year, Month, Weekday and Day and seeing how the .at_datetime() is useful.
.drop() can be used later to remove them.

```python
df['Year'] = pd.DatetimeIndex(df.date).year
```

```python
df
```

Is'nt that cool!!

```python
df['Month'] = pd.DatetimeIndex(df.date).month
df['Weekday'] = pd.DatetimeIndex(df.date).weekday
df['Day'] = pd.DatetimeIndex(df.date).day
```

```python
df
```

This above is how a dataframe should be converted and broken down for a better time series analysis

# Group By and Aggregation

Step by Step process to understand groupby function

```python
#Using Groupby function based on month 
df.groupby('Month')
```

```python
#Assigning the groupby generic.DataFrameGroupBy object to a variable
df_months = df.groupby('Month')
```

```python
#Calling few columns from the assigned variable
df_months[['new_cases','new_deaths','new_tests']]
```

```python
#Printing the groupedby rows and columns
df_months[['new_cases','new_deaths','new_tests']].sum()
```

The above codes can be written in just one line of code using chaining feature available in pandas

```python
df_weekday = df.groupby('Weekday')[['new_cases','new_deaths','new_tests']].mean()
```

# Doing a cumulative sum. This is very important for time series data

```python
df['total_cases'] = df.new_cases.cumsum()
```

```python
df['total_deaths'] = df.new_deaths.cumsum()
```

```python
initial_tst = 935310
df['total_tests'] = df.new_tests.cumsum()+initial_tst
```

# Merging Data from Multiple Sources

```python
urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/locations.csv', 
            'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location=='Italy']
```

```python
df['location'] = 'Italy'
```

```python
df
```

The above step is done so as to have a common column between dataframe df and df_location

Using merge() method to join two dataframes

```python
merged_df = df.merge(locations_df,on="location")
```

```python
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases*1e6 /merged_df.population
```

Writing data back to files

```python
final_df = merged_df[['date','new_cases','total_cases','new_deaths','total_deaths','new_tests','total_tests','cases_per_million']]
```

```python
final_df
```

```python
final_df.to_csv('analysis.csv',index=None)
```
